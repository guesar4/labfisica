```python
import matplotlib.pyplot as plt
import math
import numpy as np
import statistics as stat
```

```python
%matplotlib inline
```

```python
corrida_1 = [0.4,0.34,0.47,0.4,0.4,0.4,0.4,0.4,0.47,0.4]

avg_1 = stat.mean(corrida_1)
stdev_1 = stat.stdev(corrida_1)
inc_1 = stdev_1/math.sqrt(len(corrida_1))
real_1 = math.sqrt(2*(82.0*0.01)/9.75589)
max_1 = stdev_1 + avg_1
min_1 = avg_1 - stdev_1
imax_1 = inc_1 + avg_1
imin_1 = avg_1 - inc_1
print(avg_1,real_1,max_1,min_1,imax_1,imin_1)
```

```python
plt.figure(dpi = 1200)
plt.hist(corrida_1,
         bins=3,
         density=True,
         histtype='bar',
         color='b',
         edgecolor='k',
         alpha=0.1)

plt.xlabel('Tiempo de caída')
plt.ylabel('Repeticiones')

plt.axvline(avg_1, color='red', linestyle='dashed', linewidth=1)
plt.text(avg_1*0.94, 16*0.9, 'Promedio: {:.2f}'.format(avg_1), fontsize='x-small', color='red')

plt.axvline(real_1, color='limegreen', linestyle='dashed', linewidth=1)
plt.text(real_1*1.03, 16*0.9, 'Valor teórico: {:.2f}'.format(real_1), fontsize='x-small', color='limegreen')

plt.axvline(max_1, color='blue', linestyle='dashed', linewidth=1)
plt.text(max_1*1.005, 12*0.9, 'Max: {:.2f}'.format(max_1), fontsize='x-small', color='blue')

plt.axvline(min_1, color='blue', linestyle='dashed', linewidth=1)
plt.text(min_1*0.92, 12*0.9, 'Min: {:.2f}'.format(min_1), fontsize='x-small', color='blue')

plt.axvline(imax_1, color='purple', linestyle='dashed', linewidth=1)
plt.text(imax_1*1.005, 10*0.9, 'Imax: {:.2f}'.format(imax_1), fontsize='x-small', color='purple')

plt.axvline(imin_1, color='purple', linestyle='dashed', linewidth=1)
plt.text(imin_1*0.94, 10*0.9, 'Imin: {:.2f}'.format(imin_1), fontsize='x-small', color='purple')


plt.show()
```
